```python
import datasets
import glob
import random
```

```python
SEED=12345
data_files = []

def add_jsonlines_dir(path):
    global data_files
    data_files += glob.glob(f"{path}/*.json.gz")
    print(f"Number of files {len(data_files)} after adding {path}")
    
add_jsonlines_dir("/home/yeb/Developer/data/nrc_uniq_cleaned_20210223")
add_jsonlines_dir("/home/yeb/Developer/data/nu_uniq_cleaned_20210225")

random.Random(SEED).shuffle(data_files)
print(data_files)
```

```python
total=len(data_files)
print(total)
tenperc=int(total/10)
nineperc=total-tenperc
print(f"90%: {nineperc}")
train=data_files[:nineperc]
test=data_files[nineperc:]
```

```python
assert list(set(train) & set(test)) == [], "Train overlaps with test"
```

```python
dataset = datasets.load_dataset('json', data_files={'train': train, 'test': test})
```
